```python
# Library

import tkinter as tk
from tkinter import ttk
import pandas as pd


# App window
win = tk.Tk()
win.title('GUI')  # Window Title

# Creating the Label and entry box for 'gender' column.

gender=ttk.Label(win,text="Gender")
gender.grid(row=0,column=0,sticky=tk.W)

gender_var=tk.StringVar()
gender_box=ttk.Combobox(win,width=13,textvariable=gender_var)
gender_box['values']=('Male','Female')
gender_box.grid(row=0,column=1)


# Creating the Label and entry box for 'Senior Citizen' column.

SR_Citizen=ttk.Label(win,text="Senior Citizen")
SR_Citizen.grid(row=1,column=0,sticky=tk.W)

SR_Citizen_var=tk.StringVar()
SR_Citizen_box=ttk.Combobox(win,width=13,textvariable=SR_Citizen_var)
SR_Citizen_box['values']=('1','0')
SR_Citizen_box.grid(row=1,column=1)


# Creating the Label and entry box for 'Partner' column.

partner=ttk.Label(win,text="Partner")
partner.grid(row=2,column=0,sticky=tk.W)

partner_var=tk.StringVar()
partner_box=ttk.Combobox(win,width=13,textvariable=partner_var)
partner_box['values']=('Yes','No')
partner_box.grid(row=2,column=1)


# Creating the Label and entry box for 'Dependents' column.

dependents=ttk.Label(win,text="Dependents")
dependents.grid(row=3,column=0,sticky=tk.W)

dependents_var=tk.StringVar()
dependents_box=ttk.Combobox(win,width=13,textvariable=dependents_var)
dependents_box['values']=('Yes','No')
dependents_box.grid(row=3,column=1)

# Creating the Label and entry box for 'Tenure' column.

tenure=ttk.Label(win,text="Tenure")
tenure.grid(row=4,column=0,sticky=tk.W)

tenure_var=tk.StringVar()
tenure_entrybox=ttk.Entry(win,width=16,textvariable=tenure_var)
tenure_entrybox.grid(row=4,column=1)


# Creating the Label and entry box for 'Phone Services' column.

phoneServices=ttk.Label(win,text="Phone Services")
phoneServices.grid(row=5,column=0,sticky=tk.W)

phoneServices_var=tk.StringVar()
phoneServices_box=ttk.Combobox(win,width=13,textvariable=phoneServices_var)
phoneServices_box['values']=('Yes','No')
phoneServices_box.grid(row=5,column=1)



# Creating the Label and entry box for 'Multiple Lines' column.

multipleLines=ttk.Label(win,text="Multiple Lines")
multipleLines.grid(row=6,column=0,sticky=tk.W)

multipleLines_var=tk.StringVar()
multipleLines_box=ttk.Combobox(win,width=13,textvariable=multipleLines_var)
multipleLines_box['values']=('Yes','No','No phone service')
multipleLines_box.grid(row=6,column=1)


# Creating the Label and entry box for 'Internet Service' column.

internetService=ttk.Label(win,text="Internet Service")
internetService.grid(row=7,column=0,sticky=tk.W)

internetService_var=tk.StringVar()
internetService_box=ttk.Combobox(win,width=13,textvariable=internetService_var)
internetService_box['values']=('DSL','Fibre Optic','No')
internetService_box.grid(row=7,column=1)


# Creating the Label and entry box for 'Online Security' column.

OnlineSecurity=ttk.Label(win,text="Online Security")
OnlineSecurity.grid(row=8,column=0,sticky=tk.W)

OnlineSecurity_var=tk.StringVar()
OnlineSecurity_box=ttk.Combobox(win,width=13,textvariable=OnlineSecurity_var)
OnlineSecurity_box['values']=('Yes','No','No Internet service')
OnlineSecurity_box.grid(row=8,column=1)


# Creating the Label and entry box for 'Online Backup' column.

OnlineBackup=ttk.Label(win,text="Online Backup")
OnlineBackup.grid(row=9,column=0,sticky=tk.W)

OnlineBackup_var=tk.StringVar()
OnlineBackup_box=ttk.Combobox(win,width=13,textvariable=OnlineBackup_var)
OnlineBackup_box['values']=('Yes','No','No Internet service')
OnlineBackup_box.grid(row=9,column=1)


# Creating the Label and entry box for 'Device Protection' column

DeviceProtection=ttk.Label(win,text="Device Protection")
DeviceProtection.grid(row=10,column=0,sticky=tk.W)

DeviceProtection_var=tk.StringVar()
DeviceProtection_box=ttk.Combobox(win,width=13,textvariable=DeviceProtection_var)
DeviceProtection_box['values']=('Yes','No','No Internet service')
DeviceProtection_box.grid(row=10,column=1)


# Creating the Label and entry box for 'Tech Support' column

TechSupport=ttk.Label(win,text="Tech Support")
TechSupport.grid(row=11,column=0,sticky=tk.W)

TechSupport_var=tk.StringVar()
TechSupport_box=ttk.Combobox(win,width=13,textvariable=TechSupport_var)
TechSupport_box['values']=('Yes','No','No Internet service')
TechSupport_box.grid(row=11,column=1)


# Creating the Label and entry box for 'Streaming TV' column.

StreamingTV=ttk.Label(win,text="Streaming TV")
StreamingTV.grid(row=12,column=0,sticky=tk.W)

StreamingTV_var=tk.StringVar()
StreamingTV_box=ttk.Combobox(win,width=13,textvariable=StreamingTV_var)
StreamingTV_box['values']=('Yes','No','No Internet service')
StreamingTV_box.grid(row=12,column=1)


# Creating the Label and entry box for 'Streaming Movies' column.

StreamingMovies=ttk.Label(win,text="Streaming Movies")
StreamingMovies.grid(row=13,column=0,sticky=tk.W)

StreamingMovies_var=tk.StringVar()
StreamingMovies_box=ttk.Combobox(win,width=13,textvariable=StreamingMovies_var)
StreamingMovies_box['values']=('Yes','No','No Internet service')
StreamingMovies_box.grid(row=13,column=1)


# Creating the Label and entry box for 'Contract' column.

Contract=ttk.Label(win,text="Contract")
Contract.grid(row=14,column=0,sticky=tk.W)

Contract_var=tk.StringVar()
Contract_box=ttk.Combobox(win,width=13,textvariable=Contract_var)
Contract_box['values']=('Month-to-month','One year','Two year')
Contract_box.grid(row=14,column=1)


# Creating the Label and entry box for 'Paperless Billing' column.

PaperlessBilling=ttk.Label(win,text="Paperless Billing")
PaperlessBilling.grid(row=15,column=0,sticky=tk.W)

PaperlessBilling_var=tk.StringVar()
PaperlessBilling_box=ttk.Combobox(win,width=13,textvariable=PaperlessBilling_var)
PaperlessBilling_box['values']=('Yes','No')
PaperlessBilling_box.grid(row=15,column=1)


# Creating the Label and entry box for 'Payment Method' column.

PaymentMethod=ttk.Label(win,text="Payment Method")
PaymentMethod.grid(row=16,column=0,sticky=tk.W)

PaymentMethod_var=tk.StringVar()
PaymentMethod_box=ttk.Combobox(win,width=13,textvariable=PaymentMethod_var)
PaymentMethod_box['values']=('Bank transfer','Credit card','Electronic check','Mailed check')
PaymentMethod_box.grid(row=16,column=1)


# Creating the Label and entry box for 'Monthly Charges' column.

MonthlyCharges=ttk.Label(win,text="Monthly Charges")
MonthlyCharges.grid(row=17,column=0,sticky=tk.W)

MonthlyCharges_var=tk.StringVar()
MonthlyCharges_box=ttk.Entry(win,width=16,textvariable=MonthlyCharges_var)
MonthlyCharges_box.grid(row=17,column=1)


# Creating the Label and entry box for 'Total Charges' column.

TotalCharges=ttk.Label(win,text="Total Charges")
TotalCharges.grid(row=18,column=0,sticky=tk.W)

TotalCharges_var=tk.StringVar()
TotalCharges_box=ttk.Entry(win,width=16,textvariable=TotalCharges_var)
TotalCharges_box.grid(row=18,column=1)

# Creating the pandas DataFrame
import pandas as pd
DF = pd.DataFrame()
    


# Creating a function for the predict button to get the final result.
def action():
    global DB
    import pandas as pd
    DF = pd.DataFrame(columns=['Gender','Senior Citizen','Partner','Dependents','Tenure','Phone Services',
                               'Multiple Lines','Internet Services','Online Security','Online Backup',
                               'Device Protection','Tech Support','Streaming TV','Streaming Movies','Contract',
                              'Paperless Billing', 'Payment Method','Monthly Charges','Total Charges'])
    
    # Gender
    gender=gender_var.get()
    if gender=='Female':
        DF.loc[0,'Gender']=0
    elif gender=='Male':
        DF.loc[0,'Gender']=1
  

    # Senior Citizen 
    SC=SrCitizen_var.get()
    DF.loc[0,'Senior Citizen']=SC

    # Partner
    partner=partner_var.get()
    if partner=='No':
        DF.loc[0,'Partner']=0
    elif partner=='Yes':
        DF.loc[0,'Partner']=1
    
    # Dependents
    dependent=dependents_var.get()
    if dependent=='No':
        DF.loc[0,'Dependents']=0
    elif dependent=='Yes':
        DF.loc[0,'Dependents']=1
    
    # Tenure
    tenure=tenure_var.get()
    DF.loc[0,'Tenure']=tenure
    
    # Phone Services
    phoneser=phoneServices_var.get()
    if phoneser=='No':
        DF.loc[0,'Phone Services']=0
    elif phoneser=='Yes':
        DF.loc[0,'Phone Services']=1
    
    # Multiple Lines 
    multilines=multipleLines_var.get()
    if multilines=='No':
        DF.loc[0,'Multiple Lines']=0
    elif multilines=='No phone service':
        DF.loc[0,'Multiple Lines']=1
    elif multilines=='Yes':
        DF.loc[0,'Multiple Lines']=2
    
    # Internet Services 
    interser=internetService_var.get()
    if interser=='DSL':
        DF.loc[0,'Internet Services']=0
    elif interser=='Fibre Optic':
        DF.loc[0,'Internet Services']=1
    elif interser=='No':
        DF.loc[0,'Internet Services']=2
    
    
    # Online Security 
    onlineser=OnlineSecurity_var.get()
    if onlineser=='No':
        DF.loc[0,'Online Security']=0
    elif onlineser=='No Internet service':
        DF.loc[0,'Online Security']=1
    elif onlineser=='Yes':
        DF.loc[0,'Online Security']=2
        
        
    # Online Backup 
    onlinback=OnlineBackup_var.get()
    if onlinback=='No':
        DF.loc[0,'Online Backup']=0
    elif onlinback=='No Internet service':
        DF.loc[0,'Online Backup']=1
    elif onlinback=='Yes':
        DF.loc[0,'Online Backup']=2
    
    
    # Device Protection 
    devicepro=DeviceProtection_var.get()
    if devicepro=='No':
        DF.loc[0,'Device Protection']=0
    elif devicepro=='No Internet service':
        DF.loc[0,'Device Protection']=1
    elif devicepro=='Yes':
        DF.loc[0,'Device Protection']=2
    
    # Tech Support
    techsup=TechSupport_var.get()
    if techsup=='No':
        DF.loc[0,'Tech Support']=0
    elif techsup=='No Internet service':
        DF.loc[0,'Tech Support']=1
    elif techsup=='Yes':
        DF.loc[0,'Tech Support']=2
    
    # Streaming TV
    steamtv=StreamingTV_var.get()
    if steamtv=='No':
        DF.loc[0,'Streaming TV']=0
    elif steamtv=='No Internet service':
        DF.loc[0,'Streaming TV']=1
    elif steamtv=='Yes':
        DF.loc[0,'Streaming TV']=2
    
    
    # Streaming Movies
    steammovi=StreamingMovies_var.get()
    if steammovi=='No':
        DF.loc[0,'Streaming Movies']=0
    elif steammovi=='No Internet service':
        DF.loc[0,'TStreaming Movies']=1
    elif steammovi=='Yes':
        DF.loc[0,'Streaming Movies']=2
    
    
    # Contract
    contract=Contract_var.get()
    if contract=='Month-to-month':
        DF.loc[0,'Contract']=0
    elif contract=='One year':
        DF.loc[0,'Contract']=1
    elif contract=='Two year':
        DF.loc[0,'Contract']=2
    
    
    # Paperless Billing
    paperles=PaperlessBilling_var.get()
    if paperles=='No':
        DF.loc[0,'Paperless Billing']=0
    elif paperles=='Yes':
        DF.loc[0,'Paperless Billing']=1
    
    
    # Payment Method
    paymethod=PaymentMethod_var.get()
    if paymethod=='Bank transfer':
        DF.loc[0,'Payment Method']=0
    elif paymethod=='Credit card':
        DF.loc[0,'Payment Method']=1
    elif paymethod=='Electronic check':
        DF.loc[0,'Payment Method']=2
    elif paymethod=='Mailed check':
        DF.loc[0,'Payment Method']=3
    
    
    # Monthly Charges
    moncharges=MonthlyCharges_var.get()
    DF.loc[0,'Monthly Charges']=moncharges
    
    # Total Charges
    tolcharges=TotalCharges_var.get()
    DF.loc[0,'Total Charges']=tolcharges
    
    
    print(DF.shape)
    DB=DF
    

Predict_button=ttk.Button(win,text="Submit",command=action)
Predict_button.grid(row=19,column=0)


# Creating a function to convert the classes of the categorical features to numeric.
def Output():
    DB["Gender"] = pd.to_numeric(DB ["Gender"]) # Converting categorical to numeric for all the features.........
    DB["Senior Citizen"] = pd.to_numeric(DB ["Senior Citizen"])
    DB["Partner"] = pd.to_numeric(DB ["Partner"])
    DB["Dependents"] = pd.to_numeric(DB ["Dependents"])
    DB["Tenure"] = pd.to_numeric(DB ["Tenure"])
    DB["Phone Services"] = pd.to_numeric(DB ["Phone Services"])
    DB["Multiple Lines"] = pd.to_numeric(DB["Multiple Lines"])
    DB["Internet Services"] = pd.to_numeric(DB["Internet Services"])
    DB["Online Security"] = pd.to_numeric(DB["Online Security"])
    DB["Online Backup"] = pd.to_numeric(DB["Online Backup"])
    DB["Device Protection"] = pd.to_numeric(DB["Device Protection"])
    DB["Online Security"] = pd.to_numeric(DB["Online Security"])
    DB["Streaming TV"] = pd.to_numeric(DB["Streaming TV"])
    DB["Streaming Movies"] = pd.to_numeric(DB["Streaming Movies"])
    DB["Contract"] = pd.to_numeric(DB["Contract"])
    DB["Paperless Billing"] = pd.to_numeric(DB["Paperless Billing"])
    DB["Payment Method"] = pd.to_numeric(DB["Payment Method"])
    DB["Monthly Charges"] = pd.to_numeric(DB["Monthly Charges"])
    DB["Total Charges"] = pd.to_numeric(DB["Total Charges"])
    
    # Importing the pickel library for load the model for prediction
    import pickle
    loaded_model = pickle.load(open('ADA Boosting.sav', 'rb')) # loading the ADA Boosting Model
    output=loaded_model.predict(DB)
    if output==1:
        result='Yes'
    elif output==0:
        result='No'
    # Prediction BOX.
    Predict_entrybox=ttk.Entry(win,width=16)
    Predict_entrybox.grid(row=20,column=1)
    Predict_entrybox.insert(1,str(result))


Predict_button=ttk.Button(win,text="Predict",command=Output) # Button for prediction.
Predict_button.grid(row=20,column=0)

# Closing the loop for final run.
win.mainloop()
```
