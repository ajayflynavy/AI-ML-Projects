```python
# Library

import tkinter as tk
from tkinter import ttk
import pandas as pd
import re


# App window
win = tk.Tk()
win.title('Nueral Networks GUI - Great Learning')  # Window Title

# Step 1: Import data frame name ---------------------------------

Name=ttk.Label(win,text="Step 1: File Name")
Name.grid(row=0,column=0,sticky=tk.W)

Name_var=tk.StringVar()
Name_entrybox=ttk.Entry(win,width=16,textvariable=Name_var)
Name_entrybox.grid(row=0,column=1)

def Import_Data():
    global DB
    DF_Name=Name_var.get()
    DB_extension=re.findall("\..*", DF_Name) 
    if DB_extension==['.xlsx']:
        DB=pd.read_excel(DF_Name)
    elif DB_extension==['.csv']:
        DB=pd.read_csv(DF_Name)
    # Blank empty window to print confirmation
    confirm="Done"
    Confirm_entrybox=ttk.Entry(win,width=16)
    Confirm_entrybox.grid(row=0,column=3)
    Confirm_entrybox.insert(1,str(confirm))   

Import_Data_Button=ttk.Button(win,text="Import Data",command=Import_Data)
Import_Data_Button.grid(row=0,column=2)

# Step 2: Target data frame name ---------------------------------

Target=ttk.Label(win,text="Step 2: Target Colummn")
Target.grid(row=1,column=0,sticky=tk.W)

Target_var=tk.StringVar()
Target_entrybox=ttk.Entry(win,width=16,textvariable=Target_var)
Target_entrybox.grid(row=1,column=1)

def Target_Data():
    global DB,X,y, Target_Name, X_train, X_test, y_train, y_test
    Target_Name=Target_var.get()
    
    Column_name=DB.columns
    Column_name
    found=0

    for i in range(len(Column_name)):
        if Column_name[i]==Target_Name:
            confirm="Found"
            y = DB[Target_Name]                # Target variable
            X = DB.drop(Target_Name,axis=1)    # Predictor variable
            # Train and text split
            from sklearn.model_selection import train_test_split
            X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=9,shuffle=True)


        else:
            confirm="Not Found"
    
    Confirm_entrybox=ttk.Entry(win,width=16)
    Confirm_entrybox.grid(row=1,column=3)
    Confirm_entrybox.insert(1,str(confirm))


Target_Button=ttk.Button(win,text="IMPORT DATA",command=Target_Data)
Target_Button.grid(row=1,column=2)

# Step 3: Modelling ---------------------------------

Modelling=ttk.Label(win,text="Step 3: Neural Network Regressor")
Modelling.grid(row=2,column=0,sticky=tk.W)


# 1. Regression
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, LeakyRelu
from tensorflow.keras import regularizers, optimizers
from keras import optimizers
from sklearn.metrics import r2_score

NN_R=ttk.Label(win,text="Regression")
NN_R.grid(row=3,column=0,sticky=tk.E)

def NN_reg_model():
    
    global DB, X, y, model_reg
    
    X = DB.drop('Signal_Strength', 1)
    y = DB.pop('Signal_Strength')
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=1)
    
    # Scaling the data
    X_train_s = StandardScaler().fit_transform(X_train)
    X_test_s = StandardScaler().transform(X_test)
    
    
    # Neural Network Regressor
    # Initialize Sequential model
    model_reg = Sequential()
    model_reg.add(Dense(128, kernel_initializer='normal', input_shape= X_train.shape[1], activation='relu'))
    model_reg.add(Dense(64, kernel_initializer='normal', activation='relu'))
    model_reg.add(Dense(32, kernel_initializer='normal'))
    model_reg.add(Dense(LeakyRelu(alpha = 0.1))
    model_reg.add(Dense(16, kernel_initializer='normal'))
    model_reg.add(Dense(LeakyRelu(alpha = 0.1)))
    model_reg.add(Dense(1, kernel_initializer='normal'))
    model_reg.add(Dense(LeakyRelu(alpha = 0.1))
    model_reg.compile(optimizer='adam', loss = 'mean_absolute_error', mertics=['accuracy'])
    model_reg.summary()
    # Fitting the training data to the model
    model_reg.fit(X_train_s, y_train, epochs=400, batch_size=200, validation_data =(X_test_s, y_test))
    
      
    Confirm_entrybox=ttk.Entry(win,width=16)
    Confirm_entrybox.grid(row=3,column=2)
    Confirm_entrybox.insert(1,str("Network Trained"))
    
NNR_Button=ttk.Button(win,text="Train",command= NN_reg_model)
NNR_Button.grid(row=3,column=1)
                  
from keras.models import model_from_json
import numpy
import os

pickel =ttk.Label(win,text="Pickle")
pickel.grid(row=4,column=0,sticky=tk.E)

def PICKLE():
    Regressor_model_json = model_reg
    with open('Regressor_model_json', 'w') as json_file:
        json_file.write(Regressor_model_json)
    
    model_reg.save_weights('Regressor_model.h5')
                  
    confirm_entrybox= ttk.Entry(win, width=16)
    confirm_entrybox.grid(row =4, column =2)
    confirm_entrybox.insert(1, str('save model to disk'))
    
pickel_Button=ttk.Button(win,text="Run",command=PICKLE)
pickel_Button.grid(row=4,column=1)

# Step 4: Modelling ---------------------------------

Modelling=ttk.Label(win,text="Step 4: Neural Network Classifier")
Modelling.grid(row=5,column=0,sticky=tk.W)

# 1. Classifier
NN_R=ttk.Label(win,text="Classifier")
NN_R.grid(row=6,column=0,sticky=tk.E)

def NN_class_model():
    global DB, X, y, model_class
    
    X = DB.drop('Signal_Strength', 1)
    y = DB.pop('Signal_Strength')
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=1)
    
    # Scaling the data
    X_train_s = StandardScaler().fit_transform(X_train)
    X_test_s = StandardScaler().transform(X_test)
    
    
    # Neural Network Regressor
    # Initialize Sequential model
    model_class = Sequential()
    # Normalizing the input Data
    model_class.add(Dense(128, kernel_initializer='normal', input_shape= X_train.shape[1], activation='relu'))
    model_class.add(Dense(64, kernel_initializer='normal', activation='relu'))
    model_class.add(Dense(32, kernel_initializer='normal',activation='relu'))
    model_class.add(Dense(16, kernel_initializer='normal',activation='relu'))
    model_class.add(Dense(9, kernel_initializer='normal',activation='softmax')
    model_class.compile(optimizer='adam', loss = 'categorical_crossentropy', mertics=['accuracy'])
    model_class.summary()
    # Fitting the training data to the model
    model_class.fit(X_train_s, y_train, epochs=400, batch_size=200, validation_data =(X_test_s, y_test))
    
      
    Confirm_entrybox=ttk.Entry(win,width=16)
    Confirm_entrybox.grid(row=6,column=2)
    Confirm_entrybox.insert(1,str("Network Trained"))
    
NNR_Button=ttk.Button(win,text="Train",command= NN_class_model)
NNR_Button.grid(row=6,column=1)
                  
from keras.models import model_from_json
import numpy
import os

pickel =ttk.Label(win,text="Pickle")
pickel.grid(row=7,column=0,sticky=tk.E)

def PICKLE():
    Regressor_model_json = model_class
    with open('Class_model_json', 'w') as json_file:
        json_file.write(Regressor_model_json)
    
    model_reg.save_weights('Class_model.h5')
                  
    confirm_entrybox= ttk.Entry(win, width=16)
    confirm_entrybox.grid(row =7, column =2)
    confirm_entrybox.insert(1, str('save model to disk'))
    
pickel_Button=ttk.Button(win,text="Run",command=PICKLE)
pickel_Button.grid(row=7,column=1)

    
win.mainloop()
```
